```python
from random import shuffle
from dataclasses import dataclass
import numpy as np

COLORS = ["green", "red", "yellow", "blue"]
VALUES = range(0,15)

WIZARD = 14
JOKER = 0

@dataclass
class Card:
    color: str
    value: int

class CardDeck:
    def __init__(self, ordered=False):
        self.cards = []
        if ordered == True:
            self.initialize_ordered_deck()
        else:
            self.initialize_deck()

    def initialize_ordered_deck(self):
        for color in COLORS:
            for value in VALUES:
                self.cards.append(Card(value, color))
    
    def initialize_deck(self):
        for color in COLORS:
            for value in VALUES:
                self.cards.append(Card(value, color))
        shuffle(self.cards)

deck = CardDeck(True)
hand = [Card(1, "green"), Card(1, "red"), Card(14, "red")]
```

```python
from cards import TRUMP

print(TRUMP)

def trump_enc(trump):
    trump_enc = np.zeros(5)         # ['green', 'red', 'blue', 'yellow', 'none']
    trump_enc[TRUMP.index(trump)] = 1.0
    return trump_enc

```

```python
multi_hot_hand = np.zeros(60)

for hand_card in hand:
    for i, deck_card in enumerate(deck.cards):
        if hand_card == deck_card:
            multi_hot_hand[i] = 1

print(multi_hot_hand)
```
